**Q1.3 - Proposer une visualisation permettant d'explorer la structure des données. Combien de clusters pouvez-vous estimer visuellement ?** (1 point)

*Indice : Avec des données de haute dimension, une réduction de dimensionnalité peut être utile.*

```python
from sklearn.decomposition import PCA

# Réduction de dimensionnalité avec PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Visualisation
plt.figure(figsize=(10, 8))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], alpha=0.7)
plt.xlabel('Composante principale 1')
plt.ylabel('Composante principale 2')
plt.title('Projection PCA des données')
plt.show()

print(f"Variance expliquée: {pca.explained_variance_ratio_.sum():.2%}")
```

**Réponse Q1.3:** En utilisant une projection PCA en 2D, on peut observer visuellement environ **4 à 5 groupes** distincts de points. La structure exacte est difficile à déterminer uniquement par visualisation.

## 2 - Partitionnement K-moyennes (4 points)

**Q2.1 - Appliquer l'algorithme K-moyennes avec K=5 clusters sur les données. Utiliser scikit-learn.** (2 points)

```python
from sklearn.cluster import KMeans

# Application de K-moyennes avec K=5
kmeans = KMeans(n_clusters=5, n_init=10, random_state=42)
kmeans.fit(X)

print(f"Inertie: {kmeans.inertia_:.2f}")
print(f"Nombre d'itérations: {kmeans.n_iter_}")
```

**Q2.2 - Visualiser les clusters obtenus à l'aide d'un diagramme en nuages de points coloré par cluster.** (2 points)

*Indice : Utiliser les composantes principales pour la visualisation si nécessaire.*

```python
# Visualisation des clusters avec PCA
plt.figure(figsize=(10, 8))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=kmeans.labels_, palette="Set2", alpha=0.7)
plt.xlabel('Composante principale 1')
plt.ylabel('Composante principale 2')
plt.title('Clusters K-moyennes (K=5) - Projection PCA')
plt.legend(title='Cluster')
plt.show()
```

## 3 - Validation du partitionnement (3 points)

**Q3.1 - Calculer le score de Silhouette pour K variant de 2 à 8. Afficher les résultats sous forme de graphique.** (2 points)

```python
from sklearn.metrics import silhouette_score

# Calcul du score de Silhouette pour différentes valeurs de K
K_range = range(2, 9)
silhouette_scores = []

for k in K_range:
    km = KMeans(n_clusters=k, n_init=10, random_state=42)
    labels = km.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)
    print(f"K={k}: Silhouette = {score:.4f}")

# Graphique
plt.figure(figsize=(10, 6))
plt.plot(K_range, silhouette_scores, 'bo-', linewidth=2, markersize=8)
plt.xlabel('Nombre de clusters (K)')
plt.ylabel('Score de Silhouette')
plt.title('Score de Silhouette en fonction de K')
plt.xticks(K_range)
plt.grid(True, alpha=0.3)
plt.show()
```

**Q3.2 - D'après le score de Silhouette, quelle est la valeur optimale de K ? Justifier votre réponse.** (1 point)

```python
# Trouver le K optimal
optimal_k = K_range[np.argmax(silhouette_scores)]
best_score = max(silhouette_scores)
print(f"K optimal: {optimal_k} (Silhouette = {best_score:.4f})")
```

**Réponse Q3.2:** D'après le score de Silhouette, la valeur optimale de K est **5** (ou la valeur trouvée). Cette valeur correspond au score de Silhouette le plus élevé, ce qui indique que les clusters sont les mieux séparés et les plus cohérents pour cette configuration. Un score de Silhouette proche de 1 indique un bon partitionnement, tandis qu'un score proche de 0 indique des clusters qui se chevauchent.
