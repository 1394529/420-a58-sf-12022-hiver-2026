### Question 1-2 - Quel est le support de la paire { Jet Li, Sylvester Stallone } ? (2 points)

### ----- Votre réponse ci-dessous -----

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

data = pd.read_csv('../../data/action-movies.csv')
transactions = list(data["cast"].apply(lambda x:x.split(',')))

te = TransactionEncoder()
te_data = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_data,columns=te.columns_)

freq = apriori(df, min_support=0.1, use_colnames=True)
freq
```

```python
freq[freq["itemsets"].apply(lambda s: "Sylvester Stallone" in s and "Jet Li" in s)]
```

 ### Question 1-3 - Considérant un seuil de support de 0.10, combien d'itemsets sont fréquents ? (2 points)

```python
len(freq)
```

### Question 1-4 - Toujours en considérant un seuil de support de 0.10, quel est l'indice de confiance de la règle d'association { Jet Li } => { Sylvester Stallone } ? (2 points)

### ----- Votre réponse ci-dessous -----

```python
# Calcul de l'indice de confiance pour la règle { Jet Li } => { Sylvester Stallone }
from mlxtend.frequent_patterns import association_rules

rules = association_rules(freq, metric="confidence", min_threshold=0.10)
rule = rules[(rules['antecedents'] == frozenset({'Jet Li'})) & (rules['consequents'] == frozenset({'Sylvester Stallone'}))]

# Réponse rédactionnelle
rule['confidence'].values[0] if not rule.empty else "Règle non trouvée"
```
